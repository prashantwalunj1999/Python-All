```python
import pandas as pd
import numpy as np

# data frame 1
d1={'Customer_id':pd.Series([1,2,3,4,5,6]),
   'Product':pd.Series(['Oven','Oven','Oven','Television','Television','Television'])}

df1=pd.DataFrame(d1)

#data frame 2

d2={'Customer_id':pd.Series([2,4,6,8]),
   'State':pd.Series(['California','California','Texas','Texas'])}

df2=pd.DataFrame(d2)
```

# Inner Join Pandas


```python
# inner join in python pandas
pd.merge(df1,df2, on='Customer_id', how='inner')
```

# Outer join Pandas 

```python
# outer join in python pandas
pd.merge(df1,df2, on='Customer_id', how='outer')
```

# Left Outer Join Pandas
* Return all rows from the left table, and any rows with matching keys 

```python
# left join in python pandas
pd.merge(df1,df2, on='Customer_id', how='left')
```

# Right Outer Join Pandas 

```python
# right join in python pandas
pd.merge(df1,df2, on='Customer_id', how='right')
```

# Row bind in Python Pandas-Append 

```python
import pandas as pd
import numpy as np

# Create a Dataframe

d={
    'Name':['Alisha','Bobby','Cathrine','Mandhana','Rocky','Sebstian','Jaqluine','Rahul','David'],
'Score1':[23,34,45,67,89,90,56,54,32],
'Score2':[76,87,76,45,23,76,77,88,98],
'Score3':[44,55,66,77,33,22,11,98,45]}

df1=pd.DataFrame(d)
df1
```

```python
import pandas as pd
import numpy as np

# Create a Dataframe - df2
d={
    'Name':['girish','komal','priya'],
    'Score1':[23,34,45],
    'Score2':[76,87,76],
    'Score3':[44,55,66]}

df2=pd.DataFrame(d)
df2
```

# Method 1: Row bind or Concatenate two dataframes in 

```python
a=pd.concat([df1,df2],ignore_index=True)
a
```

# # Method 2: Row bind or Concatenate two dataframes

```python
df1.append(df2,ignore_index=True)
```

```python
import pandas as pd
import numpy as np

# data frame 1
d1={'Customer_id':[1,2,3],
    'Product':['Oven','Oven','Television'],
   'State':['California','Texas','Georgia']}

df1=pd.DataFrame(d1)
df1
```
