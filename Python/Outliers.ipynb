# Winsorizing Technique

```python
df=pd.read_excel(r"C:\Users\justw\Downloads\myworkbook.xlsx")
```

```python
sns.boxplot(y='hp',data=df)
```

```python
df.hp.describe()
```

```python
q1=df['hp'].quantile(0.25)
q1
```

```python
q1=df['hp'].quantile(0.25)
q3=df['hp'].quantile(0.75)
iqr=q3-q1
Lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
```

```python
Lower_limit,upper_limit
```

```python
df['hp']=np.where(df['hp']> upper_limit,upper_limit,df['hp'])
```

```python
df['hp']=np.where(df['hp']> Lower_limit,Lower_limit,df['hp'])
```

```python
ax=sns.boxplot(y=df["hp"])
```

```python
df['hp'].describe()
```

# To Replace with Mean, Median, Mode

```python
import pandas as pd
import numpy as np
import seaborn as sns

df=pd.read_excel(r"C:\Users\justw\Downloads\myworkbook.xlsx")
df.head()
```

```python
ax=sns.boxplot(y=df["hp"])
```

```python
from scipy import stats


mean=np.mean(df.hp)

median=np.median(df.hp)

mode=list(stats.mode(df.hp)[0])[0]
```

```python
mean,median,mode
```

```python
df.hp=np.where(df.hp > upper_limit,mean,df.hp)
```

```python
ax=sns.boxplot(y=df["hp"])
```

```python
df.hp=np.where(df.hp > upper_limit,median,df.hp)
```

```python
df.hp=np.where(df.hp > upper_limit, mode,df.hp)
```
