```python
#Quesion2


import pandas as pd
a={'List':pd.Series([1,2,3,4,5]),
               'Data':pd.Series([11,22,33,44,55,])}
df1=pd.DataFrame(a)
df1
```

```python
import pandas as pd
b={'List':pd.Series([1,2,3,4,5,6]),
    'Customer_id':pd.Series([11,22,33,44,55,]),
    'Day':pd.Series(['mon','tues','thus','wed','fri'])}
df2=pd.DataFrame(b)
df2
```

```python
pd.merge(df1, df2, on='List', how='inner')
```

```python
pd.merge(df1, df2, on='List', how='outer')
```

```python
pd.merge(df1, df2, on='List', how='left')
```

```python
pd.merge(df1, df2, on='List', how='right')
```
