```python
import pandas as pd  ## dataframe
 
# data frame 1
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6]),
      'Product':pd.Series(['Oven','Oven','Oven',
                       'Television',
                       'Television',
                       'Television'])}
df1 = pd.DataFrame(d1)
 
# data frame 2
d2 = {'Customer_id':pd.Series([2,4,6,8]),
      'State':pd.Series(['California','California','Texas','Texas'])}
df2 = pd.DataFrame(d2)
```

```python
d2
```

```python
df1
```

```python
df2
```

# Inner join pandas

* Return only the rows in which the left table have matching keys in the right table

```python
#inner join in python pandas
pd.merge(df1, df2, on='Customer_id', how='inner')
```

# Outer join pandas

*Returns all rows from both tables, join records from the left which have matching keys in the right table.

```python
# outer join in python pandas
pd.merge(df1, df2, on='Customer_id', how='outer')
```

# Left outer join pandas

* Return all rows from the left table, and any rows with matching keys from the right table.

```python
# left join in python
pd.merge(df1, df2, on='Customer_id', how='left')
```

# Right outer join pandas
Return all rows from the right table, and any rows with matching keys from the left table.

```python
# right join in python pandas
pd.merge(df1, df2, on='Customer_id', how='right')
```

# Row bind in python pandas – Append

```python
import pandas as pd
import numpy as np
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','Cathrine','Madonna',
            'Rocky','Sebastian','Jaqluine','Rahul','David'],
   'Score1':[62,47,55,74,31,77,85,63,42],
   'Score2':[89,87,67,55,47,72,76,79,44],
   'Score3':[56,86,77,45,73,62,74,89,71]}
 
df1 = pd.DataFrame(d)
df1
```

```python
import pandas as pd
import numpy as np
 
# create dataframe - df2
d = {
    'Name':['Andrew','Ajay','Teresa'],
   'Score1':[32,71,57],
   'Score2':[92,99,69],
   'Score3':[67,97,68]}
 
df2 = pd.DataFrame(d)
df2
```

# Method 1: Append or Row bind or concatenate two dataframes in pandas :
Now lets concatenate or row bind two dataframes df1 and df2

```python
a=pd.concat([df1,df2],ignore_index=True)
a
```

# Method 2: Row bind or concatenate two dataframes in pandas:

```python
df1.append(df2,ignore_index=True)
```

# Rename all the column names in python

```python
import pandas as pd
import numpy as np
 
# data frame 1
d1 = {'Customer_id':pd.Series([1,2,3]),
'Product':pd.Series(['Oven','Oven','Television']),
      'State':pd.Series(['California','Texas','Georgia'])}
df1 = pd.DataFrame(d1)
df1   
```

```python
df1.columns
```

```python
# rename all the columns in python
df1.columns = ['Customer_unique_id', 'Product_type','Province']    
```

```python
df1
```

# Rename the specific column in python:

```python
# rename province to state
df1.rename(columns={'Province': 'City'}, inplace=True)
df1
```
