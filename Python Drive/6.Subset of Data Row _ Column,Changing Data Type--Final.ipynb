# Summary of just the indexing operator


* Its primary purpose is to select columns by the column names
* Select a single column as a Series by passing the column name directly to it: df['col_name']
* Select multiple columns as a DataFrame by passing a list to it: df[['col_name1', 'col_name2']]

```python
df.head(3)
```

```python
df['mpg']  # by default it work for columns
```

```python
df[['mpg']]  # to get column name 
```

```python
df[['mpg', 'cyl', 'hp']]
```

# Getting started with .loc

The .loc indexer selects data in a different way than just the indexing operator. It can select subsets of rows or columns. It can also simultaneously select subsets of rows and columns.Most importantly, it only selects data by the LABEL of the rows and columns.

```python
df.head()
```

### [ row Label ,column Label ]

```python
df.loc[4,:]  # 1 row and : all data
```

#### Select multiple rows as a DataFrame with

```python
df.loc[0:3,:] # select 0 to 5 row and all columns
```

#### Use slice notation to select a range of rows with .loc

```python
df.loc[0:3] # without , by default take row
```

```python
df.loc[1:10,:]
```

# Changing Data Type

* There is 2 Method 

* one variable 
* more than 1 variable

```python
df.head()
```

```python
df.info() # Data Structure
```

```python
df.cyl=df.cyl.astype(str) 
# str # converting data type of 1 column
```

```python
convert_dict = {'vs': str,'am': str} 
# converting data type of more than 1 column
df = df.astype(convert_dict) 
```

```python
df.info()
```

```python
df.head()
```

```python
df.cyl.value_counts()
```

```python
a=df.loc[df.cyl=='6',:]  # Conditional Subset
a
```

```python
df.loc[df.cyl=='6','mpg':'hp']  # Conditional Subset by row n columns
```
