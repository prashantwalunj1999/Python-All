# sort a dataframe in python pandas

* Sort the python pandas Dataframe by single column – Ascending & Descending order
* Sort the pandas Dataframe by Multiple Columns
* by index

```python
import pandas as pd 
import numpy as np
 
#Create a Dictionary of series
d = {'Name':['Alisa','Bobby','Cathrine','Madonna',
             'Rocky','Sebastian','Jaqluine','Rahul',
             'David','Andrew','Ajay','Teresa'],
   'Age':[36,27,25,24,31,27,25,33,42,32,51,47],
   'Score':[89,87,67,55,47,72,76,79,44,92,99,92]}
 
#Create a DataFrame
df = pd.DataFrame(d)
df
```

# Sort the python pandas Dataframe by single column – Ascending order:

```python
df.sort_values(by='Score')
```

# Descending order

```python
# sort the pandas dataframe by descending value of
# single column
 
df.sort_values(by='Score',ascending=False)
```

# Sort the pandas Dataframe by index Columns

```python
# # sort the pandas dataframe by index ascending
df1=df.sort_index()
df1
```

```python
# sort the pandas dataframe by index descending
 
df2=df.sort_index(ascending=False)
df2
```

# Drop or delete column in python pandas

* Drop a column by name
* Drop a column based on column index
* Delete a column based on column name

# Drop a column by name:

```python
df.head()
```

```python
# drop a column based on name
 
df.drop('Age',axis=1) # axis=1 denotes column
```

```python
df.head()
```

```python
# drop Multiple  column based on name
df.drop(['Age','Score'],axis=1) # axis=1 denotes column
```

# Drop a column based on column index

```python
df.head()
```

```python
# drop a column based on column index
 
df.drop(df.columns[1],axis=1) 
# index 2 is dropped(3rd column)
```

```python
df.head()
```

```python
df=df.drop(2,axis=0) # axis=1  means column index
df
```

# Delete a column based on column name

```python
del df['Age']
df
```

# Duplicate Row

```python
#Create a DataFrame
d = {'Name':['Alisa','Bobby','jodha','jack','raghu',
            'Cathrine','Alisa','Bobby','kumar','Alisa',
             'Alex','Cathrine'],
     'Age':[36,24,23,22,23,24,
            26,24,22,23,24,24],
      
     'Score':[85,63,55,74,31,77,
              85,63,42,62,89,77]}
 
df = pd.DataFrame(d)
df
```

# Find the duplicate row in pandas

```python
df["check_duplicate"]= df.duplicated()
df
```

# Drop the duplicate rows

```python
import numpy as np
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','jodha','jack','raghu','Cathrine',
            'Alisa','Bobby','kumar','Alisa','Alex','Cathrine'],
    'Age':[36,24,23,22,23,24,26,24,22,23,24,24],
      
       'Score':[85,63,55,74,31,77,85,63,42,62,89,77]}
 
df = pd.DataFrame(d)
df
```

```python
# drop duplicate rows

df.drop_duplicates() # need to run again dataframe 
```

# Drop the duplicate by retaining last occurrence


```python
# drop duplicate rows
 
df.drop_duplicates(keep='last')
```
